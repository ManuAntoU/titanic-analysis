# STEP 3: Data Cleaning
  before analysis, we must handle all missing values


## Missing value imputation:
Imputation is process of replacing missing data with substitued value.
This stratigy depends on its data type and its distribution.

*   **Numberical data**: for skewed distribution (like "Age", "Fare"), using **median** is more robost that **mean** because median is not affected by outliers.[The mean provides an arithmetic average, while the median identifies the middle value in an ordered list]

*   **Catagorical data**: A common strategy is to fill it with **MODE**(most common used data)

*   **High Cardinality/ too many missing values**: for column like CABIN. where most data is missing, imputing might not be effective. we could either drop the column or engineer a new feature (feature engineering) from it(e.g: Has_Cabin).

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
midian = df["Age"].median()
print(midian)
```

```python
df["Age"] = df["Age"].fillna(midian)
```

```python
df.isnull().sum()
```

```python
df["Embarked"]
```

```python
mode = df["Embarked"].mode()[0]
print(mode)
```

# Now we Start Analysis Process(as dealt with cleaning of data)
## Theoretical Concepts: Univariate Analysis:
this is the simplest form of data analysis, where the data being analysed contains only one variavle. the mans purpose is to describe the data and find patterns within it.

*   **For Categorical Variables:** We use frequency tables, bar charts(countplot), or pie charts to see the count or proportion of the each category.
*   **For Numerical Variables:** We use histograms(histplot) or kernel denisty plots(kdeplot) to understand the distribution, and box plots(boxplots) to identify the central terndency, speed and outliers.




```python
print("Analyzing categorical features:")

#Set up the figure for Plotting
fig, axes = plt.subplots(2,3, figsize=(18,11))
fig.suptitle("Univsriate Analysis of \"Categotical features\"", fontsize= 16)

#Plotting each categorical features
sns.countplot(ax=axes[0,0], x='Survived', data=df).set_title('Survival Distribution')
sns.countplot(ax=axes[0,1], x='Pclass', data=df).set_title('Passenger Class Distribution')
sns.countplot(ax=axes[0,2], x='Sex', data=df).set_title("Sex Dsitribution")
sns.countplot(ax=axes[1,0], x='SibSp',data=df).set_title("Sibling Spouse Count")
sns.countplot(ax=axes[1,1], x='Parch',data=df).set_title("Parent/Children Count")
sns.countplot(ax=axes[1,2], x='Embarked',data=df).set_title("Embarkation Port")

plt.tight_layout()
plt.show()
```

Key Insights of Univariate Analysis (Categorical List):
* Survival: Over 500 passengers did not survive
* Pclass: The 3rd Class was the most populated, followed by 1st and then 2nd.
* Sex: Number of Male > Number of females
* SibSp & Parch: Most passengeres travelled alone.
* Embarked: Most passenger boarded from S, followed by C and then Q.


```python
print("Analyzing Numerical list:")

fig, axes = plt.subplots(1,2, figsize=(12,6))
fig.suptitle("Univeriate Analysis for \"Numerical List\"", fontsize=16)
#Plotting each categorical features

sns.histplot(ax=axes[0], x='Age', data=df, kde=True, bins = 30).set_title("Age Distribution")

sns.histplot(ax=axes[1], x='Fare', data=df, kde=True, bins = 30).set_title("Fare Distribution")


plt.show()
```

#Bivariate Analysis
here, we explore the relationship between two variables. Our primary focus will be on how much each feature relates to our target variable, `Survived`.



Theoretical Concepts: Bivariate Analysis

this type of analyysis involves two different variables, and its main purpose is to fins relationships between them.
* Categorical Vs Numerical: To compare a numcerical variable across different categories, we often use bar plots(`barplots`) that show the mean (or another estimator) of the numerical variavle for each category. We can also use box plots or violin plots.
* Categorical Vs Categorical: We can use stacked bar charts or contigency tables(crosstables).
* Numerical vs Numerical: A scatter plot is the standard choice, with a correlation matrix being used to quantify the relationship.

```python
print("Bivariate Analysis:")
fig, axes= plt.subplots(2,2, figsize=(18,12))
fig.suptitle("Bivariate Analysis with Survival", fontsize=16)

#Plotting each categorical features
#Pclass vs Survived
sns.barplot(ax=axes[0,0], x="Pclass", y= "Survived", data=df).set_title("Pclass vs Survived")

#Sex vs Survived
sns.barplot(ax=axes[0,1], x="Sex", y="Survived", data=df).set_title("Sex vs Survived")

#Age vs Survived
sns.barplot(ax=axes[1,0], x="Has_Cabin", y="Survived", data=df).set_title("Age vs Survived")

#Embarked vs Survived
sns.barplot(ax=axes[1,1], x="Embarked", y="Survived", data=df).set_title("Embarked vs Survived")

plt.tight_layout()
plt.show()
```

Key Insights of Bivariate Analysis with Survival:
* Plcass: A clear trend is seen here: 1st class passenger have ~= 60% survival rate, then 2nd class with 45% then finally 3r class is < 25% survival rate.

* Sex(Strongest predictor): F have > 70% survival rate and M have < 20% survival rate.

* Has_Cabin: Passenger with Cabin have higher survival rate, likely corelated with being the 1st class.

* Embarked: Passenger embarked from C have higher survival rate compared to other ports.

#step 6: Featuring Engineering

Now, we'll create new features from the existing ones to potentially uncover deeper insights and provide more uneful information for a ML model.

###Common techniques:
1. Combining Features: Creating a new feature by combining other( eg: SibSp + Parch = FamilySize).
2. Extracting from text: pulling out specific information from a text feature( eg: extracting titles from `Name` column)
3. Binning: Converting a continous numerical features into a categorical one( Binning Age into Child, Adult, Senior)


```python
#1. Create a "FamilySize" Feature:
df["FamilySize"] = df["Parch"] + df["SibSp"] + 1
```

```python
#2. Create an "IsAlone" feature:
df["isAlone"] = 0
df.loc[df["FamilySize"] == 1, "isAlone"] = 1
```

```python
df.head()
```

```python
fig, axes = plt.subplots(1,2, figsize = (12, 8))
fig.suptitle("Feature Engineerings: ")

#Survival Rate WRT Famlyszie
sns.barplot(ax=axes[0], x="FamilySize", y="Survived", data=df).set_title("Survival Rate with Family Size")

#Survival Rate WRT Alone
sns.barplot(ax=axes[1], x="isAlone", y="Survived", data=df).set_title("Survival Rate with isAlone")
```

```python
#Extract "Title" from the "Name" column
df["Title"] = df["Name"].str.extract(r' ([A-Za-z]+)\.', expand=False)

print("Extracted Titles:")
df["Title"].value_counts()
```

```python
df["Title"] = df["Title"].replace(["Lady", "Jonkheer", "Countess", "Capt", "Sir",  "Don", "Mme", "Major", "Col", "Rev", "Dr"], "Rare")

df["Title"] = df["Title"].replace("Mlle", "Miss")
df["Title"] = df["Title"].replace("Ms", "Miss")
df["Title"] = df["Title"].replace("Mme", "Rare")

#Let's see the survival rate by the now, cleaned title
plt.figure(figsize=(12,6))
sns.barplot(x="Title", y="Survived", data=df)
plt.title("Survival Rate by Title")
plt.ylabel("Survival Probability")
plt.show()
```

#Step 7: MultiVariate Analysis

Now we explore interactions between multiple variables simultaneously, including our new engineered features.


```python
sns.catplot(x="Pclass", y="Survived", hue="Sex", kind="bar", data=df, height=6, aspect=1.5)
plt.title("Survival Rate by Pclass and Sex")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x="Sex", y="Age", hue="Survived", data=df, split = True, palette={0: "blue", 1: "orange"})
plt.title("Survival Rate by Sex and Age")
plt.show()
```

Insights from Violin Plot:
* For Males, the Peak of the distribution for survivors(orange) is at a very young age(children), While peak for non-survivors is in the 20-30 range.

* For Female, the distribution of survivors is much broader, indicating that females of most ages had a good chance of surviving.


#Step 8: Correlation Analysis

```python
plt.figure(figsize=(12,6))
numeric_columns = df.select_dtypes(include=['number']).columns
correlation_matrix = df[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

Interpretaion of the Heatmap:
* `Survived` has a notable possible correlation with `Fare` and `Has_Cabin`, and a negative correlation with `Pclass` and our new   `isAlone` features.
* `Pclass` anf `fare` are strongly negatively correlated, which makes sense (1st Class = high fare)
* Our New `Familyize` is composed of `Sibsp` and `Parch`, so it's highly correlated with them.
